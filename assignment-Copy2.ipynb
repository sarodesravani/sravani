```python
import requests
from bs4 import BeautifulSoup

class Scraper:
    def __init__(self,url):
        self.final_url = url + '/2e/chapter5/'
        self.pull_data(self.final_url)
    def pull_data(self, finalurl):
        print(finalurl)
        self.result = requests.get(finalurl)
        self.soup = BeautifulSoup(self.result.content)
        self.results=self.soup.find('div',attrs={'class','calibre'})
        print(self.results.text)
        
```

```python
scraper=Scraper('https://automatetheboringstuff.com')
```
